```python
import warnings
import os
import random

warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python
import pandas as pd
import glob
from IPython.display import HTML, display
```

```python
def get_comparisons():
    comparisons = []
    captions = glob.glob('images/*')
    random.shuffle(captions)
    
    for caption in captions:
        actual = random.choice(glob.glob(caption + '/actual/*'))
        scrambled = random.choice(glob.glob(caption + '/scrambled/*'))
        comparisons.append({ 
            'caption' : f"<td style='text-align:center'>{caption[9:]}</td>", 
            'actual' : f"<td style='text-align:center'><img src='{actual}'></td>", 
            'scrambled' : f"<td style='text-align:center'><img src='blah.png'></td>", 
        })
        
    return comparisons
```

# Which is the better image generator?

8 image-generating AIs -- named **A, B, C, D, E, F, G, H** -- were asked to generate images for each text prompt. 

We want you to compare pairs of AIs (A vs B, C vs D, and so on) based on the quality of the images they generate and how well each image fits the given prompt. You will be asked to make a judgment on A vs. B using just one image from each. For the subsequent comparisons, we will give you more pairs of images so that you can make a more informed judgment.

To begin, please enter your ReaderID below and click 'SEE IMAGES' (scroll down to view all of the images). After reading, **please go back to Mechanical Turk and fill out the survey**.

```python
output = []
NUM_AGENTS = 150

for i in range(0, NUM_AGENTS):
    comparisons = list(get_comparisons())

    body = ""
    for j in range(1):
        c = comparisons.pop()
        body = body + "<tr>" + c['caption'] + c['actual'] + c['scrambled'] + "</tr>"
        
    table1 = """
    <table>
      <thead>
          <tr>
            <th style='text-align:center'>Text</th>
            <th style='text-align:center'>System A</th>
            <th style='text-align:center'>System B</th>
          </tr>
      </thead>
      <tbody>
          %s
      </tbody>
    </table>
    """ % (body)
    
    body = ""
    for j in range(2):
        c = comparisons.pop()
        body = body + "<tr>" + c['caption'] + c['actual'] + c['scrambled'] + "</tr>"
        
    table2 = """
    <table>
      <thead>
          <tr>
            <th style='text-align:center'>Text</th>
            <th style='text-align:center'>System C</th>
            <th style='text-align:center'>System D</th>
          </tr>
      </thead>
      <tbody>
          %s
      </tbody>
    </table>
    """ % (body)
    
    body = ""
    for j in range(4):
        c = comparisons.pop()
        body = body + "<tr>" + c['caption'] + c['actual'] + c['scrambled'] + "</tr>"
        
    table4 = """
    <table>
      <thead>
          <tr>
            <th style='text-align:center'>Text</th>
            <th style='text-align:center'>System E</th>
            <th style='text-align:center'>System F</th>
          </tr>
      </thead>
      <tbody>
          %s
      </tbody>
    </table>
    """ % (body)
    
    body = ""
    for j in range(8):
        c = comparisons.pop()
        body = body + "<tr>" + c['caption'] + c['actual'] + c['scrambled'] + "</tr>"
    
    table8 = """
    <table>
      <thead>
          <tr>
            <th style='text-align:center'>Text</th>
            <th style='text-align:center'>System G</th>
            <th style='text-align:center'>System H</th>
          </tr>
      </thead>
      <tbody>
          %s
      </tbody>
    </table>
    """ % (body)

    output.append(f"<div class='agent' id='agent{i}' style='display:none'>" + "<h1>A vs. B</h1>" + table1 + "<br/><h1>C vs. D</h1>" + table2 + "<br/><h1>E vs. F</h1>" + table4 + "<br/><h1>G vs. H</h1>" + table8 + "</div>")
    
display(HTML(
"""
<style>
table {
    width: 80%;
}

th {
    font-size: 1.5em;
}

td img{
    display: block;
    margin-left: auto;
    margin-right: auto;
    width: 100%;
}
</style>
"""
))

display(HTML(
"""
<div class="jp-Cell">
Enter your ReaderID: <input id="myInput" type="number" value="0" min="%s" max="%s">
<button id="myBtn" onclick="makeVisible()">SEE IMAGES</button>
</div>
""" % (0, NUM_AGENTS)
))

display(HTML("<center><div>" + ''.join(output) + "</div></center>"))

display(HTML(
"""
<script>
var input = document.getElementById("myInput");
input.addEventListener("keypress", function(event) {
  if (event.key === "Enter") {
    event.preventDefault();
    document.getElementById("myBtn").click();
  }
});

function makeVisible() {
  var val = document.getElementById("myInput").value;  
  var elems = document.getElementsByClassName('agent');
  for (var i=0;i<elems.length;i+=1){
    elems[i].style.display = 'none';
  }

  document.getElementById("agent" + val).style.display = 'block';
}
</script>
""" 
))
```
